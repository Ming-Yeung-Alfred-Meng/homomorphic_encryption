```python
import numpy as np
import random
import math
```

```python
private_key_length = 10
q_length = 3

count = 0
for i in range(1000):
    private_key = random.randrange(2 ** (private_key_length - 1), 2 ** private_key_length)
    print(f"private_key: {private_key} from [{2 ** (private_key_length - 1)}, {2 ** private_key_length})")
    
    m0 = 0
    m1 = 1
    
    q0 = random.randrange(2 ** (q_length - 1), 2 ** q_length)
    q1 = random.randrange(2 ** (q_length - 1), 2 ** q_length)
    print(f"q: {q0} from [{2 ** (q_length - 1)}, {2 ** q_length})")
    print(f"q: {q1} from [{2 ** (q_length - 1)}, {2 ** q_length})")
    
    noise0 = random.randrange(0, math.floor((-1 + math.sqrt(private_key - 1)) / 2))
    noise1 = random.randrange(0, math.floor((-1 + math.sqrt(private_key - 1)) / 2))
    print(f"noise: {noise0} from [{math.ceil((-1 - math.sqrt(private_key - 1)) / 2)}, {math.floor((-1 + math.sqrt(private_key - 1)) / 2)})")
    print(f"noise: {noise1} from [{math.ceil((-1 - math.sqrt(private_key - 1)) / 2)}, {math.floor((-1 + math.sqrt(private_key - 1)) / 2)})")
    
    c0 = private_key * q0 + 2 * noise0 + m0
    print(f"m0: {m0} -> c0: {c0}")
    c1 = private_key * q1 + 2 * noise1 + m1
    print(f"m1: {m1} -> c0: {c1}")
    
    m0_d = c0 % private_key % 2
    print(f"c0: {c0} -> m0_d: {m0_d}")
    m1_d = c1 % private_key % 2
    print(f"c1: {c1} -> m1_d: {m1_d}")
    
    added_m = (c0 + c1) % private_key % 2
    print(f"c0 + c1: {c0 + c1} -> m0_d: {added_m}")
    multiplied_m = (c0 * c1) % private_key % 2
    print(f"c0 * c1: {c0 * c1} -> m1_d: {multiplied_m}")
    
    if multiplied_m == 0:
        count += 1
        
print(count)
```
